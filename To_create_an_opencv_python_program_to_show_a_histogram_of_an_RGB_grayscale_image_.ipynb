```python
def save_and_show_histogram(image, title, folder):
    # If the image is grayscale
    if len(image.shape) == 2:
        plt.hist(image.ravel(), 256, [0,256])
        plt.title(title)
        plt.savefig(os.path.join(folder, title + '_histogram.png'))
        plt.show()
        plt.close()
    # If the image is RGB
    elif len(image.shape) == 3:
        color = ('b', 'g', 'r')
        for i,col in enumerate(color):
            hist = cv2.calcHist([image], [i], None, [256], [0,256])
            plt.plot(hist, color = col)
            plt.xlim([0,256])
        plt.title(title)
        plt.savefig(os.path.join(folder, title + '_histogram.png'))
        plt.show()
        plt.close()
    else:
        print("Unsupported image format")
```

```python
# Load image
image = cv2.imread('images.jpg')

# Convert image to grayscale
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Create a folder to save histograms
folder = 'histograms'
if not os.path.exists(folder):
    os.makedirs(folder)

# Save and show histograms
save_and_show_histogram(image, 'RGB_Histogram', folder)
save_and_show_histogram(gray_image, 'Grayscale_Histogram', folder)
```
